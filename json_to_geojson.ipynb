```python
import json
from geojson import Point, Feature, FeatureCollection, dump
import copy
import urllib.request
```

```python
dict_features = dict() 
dict_features.update({'type':'featureCollection'})

dict_geojson = dict()
dict_geojson.update({'type': 'feature'})
```

```python
f = open('/content/link.txt', 'r')
stroka = f.read()
data=json.loads(stroka)
```

```python
#link = r'https://www.saveecobot.com/storage/maps_data.js?date=2022-02-25T130'
 
#response = urllib.request.urlopen(link)

#stroka=response.read()
#data=json.loads(stroka)
```

```python
features = []
for dataList in data:

    #features = []

    coordinate1 = dataList.get("a", 0)
    coordinate2 = dataList.get("n", 0)
    
    point = Point((float(coordinate1), float(coordinate2)))
    
    dict_geojson.update({'geometry': point})
    dict_geojson.update({'properties': dataList})
    
    features.append(dict_geojson)

    dict_geojson=copy.copy(dict_geojson)


print(features)

```
